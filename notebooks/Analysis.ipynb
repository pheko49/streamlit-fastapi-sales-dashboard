## Data Cleaning
We identified duplicated rows and as such we drop the duplicated rows in order for our analysis to be distorted that could lead to incorrect decison making

```python
df1 = df.copy() # Create a copy of the dataframe to not alter the origianl dataframe

df1 = df1.drop_duplicates() # Drop the duplicated rows in the dataframe
df1
```

## Profit by Region

We analyze profitability across different regions to identify high-performaing areas

```python
df1.groupby('Region').agg({'Profit': 'sum'}).sort_values(by='Profit', ascending=False)
```

The West region generates the highest profit, while the Central region underperforms, indicating potential inefficiencies or lower demand

## Sales by Category

We examine which product categories contribute the most to overall sales

```python
df1.groupby('Category').agg({'Sales': 'sum'}).sort_values(by='Sales', ascending=False)
```

Technology contributes significantly to total sales, suggesting strong demand in that category.

## Profit vs Sales

We compare sales and profit to identify areas where high revenue does not necessarily translate into profitability. 

```python
df1.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'})
```

## Key Insights

- The West region is the most profitable, while the Central region lags behind.
- Technology products generate high revenue and strong profit margins.
- Some categories show high sales but lower profitability, indicating cost or discount issues.

These insights can help guid strategic decisions in pricing, inventory, and regional focus.
